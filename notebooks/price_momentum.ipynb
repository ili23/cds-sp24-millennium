```python
import get_data
import pandas as pd
```

```python
query = """
WITH CTE AS (SELECT 
  conm, 
  datadate, 
  prccq,
  fyearq, fqtr,
  prccq - LAG(prccq, 1) OVER (PARTITION BY conm ORDER BY fyearq, fqtr) AS price_dif
FROM quarterly)
SELECT
  conm, datadate, prccq, 
  SUM(price_dif) OVER(PARTITION BY conm ORDER BY fyearq, fqtr ROWS 3 preceding) AS twelve_month_sum 
FROM CTE"""
df = pd.read_sql_query(query, get_data.engine)
```

```python
df['twelve_month_sum'] = df['twelve_month_sum'] / df['prccq']
```
