```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

if 'tic' not in df.columns:
    print("Error: 'tic' column not found in the DataFrame.")
else:
    apple_ticker = 'AAPL'
    google_ticker = 'GOOGL'

    apple_df = df[df['tic'] == apple_ticker]
    google_df = df[df['tic'] == google_ticker]

    if apple_df.empty:
        print(f"Error: No data found for Apple ticker {apple_ticker}.")
    elif google_df.empty:
        print(f"Error: No data found for Google ticker {google_ticker}.")
    else:
        required_columns = ['datadate', 'saleq', 'oiadpq']
        missing_columns_apple = [col for col in required_columns if col not in apple_df.columns]
        missing_columns_google = [col for col in required_columns if col not in google_df.columns]

        if missing_columns_apple:
            print(f"Error: Missing columns in Apple DataFrame: {missing_columns_apple}")
        elif missing_columns_google:
            print(f"Error: Missing columns in Google DataFrame: {missing_columns_google}")
        else:
            # make datetime objects
            apple_df['datadate'] = pd.to_datetime(apple_df['datadate'])
            google_df['datadate'] = pd.to_datetime(google_df['datadate'])

            # calc operating profit margin 
            apple_df['op_margin'] = apple_df['oiadpq'] / apple_df['saleq']
            google_df['op_margin'] = google_df['oiadpq'] / google_df['saleq']

            # calc sales growth
            apple_df['sales_growth'] = apple_df['saleq'].pct_change()
            google_df['sales_growth'] = google_df['saleq'].pct_change()

            # dropping Invalid Values
            apple_df.dropna(subset=['op_margin', 'sales_growth'], inplace=True)
            google_df.dropna(subset=['op_margin', 'sales_growth'], inplace=True)

            # winsorization
            winsorization_level = 0.05  # 5% winsorization level
            apple_df['op_margin'] = stats.mstats.winsorize(apple_df['op_margin'], limits=[winsorization_level, winsorization_level])
            apple_df['sales_growth'] = stats.mstats.winsorize(apple_df['sales_growth'], limits=[winsorization_level, winsorization_level])
            google_df['op_margin'] = stats.mstats.winsorize(google_df['op_margin'], limits=[winsorization_level, winsorization_level])
            google_df['sales_growth'] = stats.mstats.winsorize(google_df['sales_growth'], limits=[winsorization_level, winsorization_level])

            # sort the DataFrames by 'datadate'
            apple_df = apple_df.sort_values('datadate')
            google_df = google_df.sort_values('datadate')

            # select the relevant columns for the new DataFrames
            apple_result = apple_df[['datadate', 'saleq', 'oiadpq', 'op_margin', 'sales_growth']]
            google_result = google_df[['datadate', 'saleq', 'oiadpq', 'op_margin', 'sales_growth']]

            print("Apple Data:")
            print(apple_result)
            print("\nGoogle Data:")
            print(google_result)

            fig, axs = plt.subplots(2, 2, figsize=(12, 8))

            axs[0, 0].hist(apple_df['op_margin'], bins=20, density=True, alpha=0.7)
            axs[0, 0].set_title('Apple - Operating Profit Margin Distribution')
            axs[0, 0].set_xlabel('Operating Profit Margin')
            axs[0, 0].set_ylabel('Density')

            axs[0, 1].hist(apple_df['sales_growth'], bins=20, density=True, alpha=0.7)
            axs[0, 1].set_title('Apple - Sales Growth Rate Distribution')
            axs[0, 1].set_xlabel('Sales Growth Rate')
            axs[0, 1].set_ylabel('Density')

            axs[1, 0].hist(google_df['op_margin'], bins=20, density=True, alpha=0.7)
            axs[1, 0].set_title('Google - Operating Profit Margin Distribution')
            axs[1, 0].set_xlabel('Operating Profit Margin')
            axs[1, 0].set_ylabel('Density')

            axs[1, 1].hist(google_df['sales_growth'], bins=20, density=True, alpha=0.7)
            axs[1, 1].set_title('Google - Sales Growth Rate Distribution')
            axs[1, 1].set_xlabel('Sales Growth Rate')
            axs[1, 1].set_ylabel('Density')

            plt.tight_layout()
            plt.show()

            # subplots for operating profit margin and sales growth rate
            fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

            ax1.plot(apple_df['datadate'], apple_df['op_margin'], label='Apple')
            ax1.plot(google_df['datadate'], google_df['op_margin'], label='Google')
            ax1.set_xlabel('Date')
            ax1.set_ylabel('Operating Profit Margin')
            ax1.set_title('Operating Profit Margin')
            ax1.legend()


            ax2.plot(apple_df['datadate'], apple_df['sales_growth'], label='Apple')
            ax2.plot(google_df['datadate'], google_df['sales_growth'], label='Google')
            ax2.set_xlabel('Date')
            ax2.set_ylabel('Sales Growth Rate')
            ax2.set_title('Sales Growth Rate')
            ax2.legend()

            plt.tight_layout()
            plt.show()
```
