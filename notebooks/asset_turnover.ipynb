```python
import pandas as pd
import numpy as np
import get_data
```

```python
df = pd.read_sql_query('SELECT datadate,conm,tic,cusip,revtq,atq FROM quarterly', get_data.engine)
clean_df = df[df[['atq', 'revtq']].notna().all(axis=1)]
clean_df['asset_turnover'] = clean_df['revtq'] / clean_df['atq']
```
